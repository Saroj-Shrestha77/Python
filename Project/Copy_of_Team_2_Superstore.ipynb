## Step 2: Data Quality Checks

Steps in data quality checks:  
1. Checking for missing values and handling them.
2. Checking for duplicates and removing them if necessary.
3. Verifying data types and converting if needed.

```python
##Checking for any null values = (Kunaipani null value chha ki bhanera janch gardai)
df_copy.isnull().sum()
```

```python
##Checking which rows have null values for postal code = (Postal code ma null value bhayeka row haru janch gardai)


df_copy[df_copy['Postal Code'].isnull()]
```

```python
##We cannot use mean or median to fill in the poastal codes =(Mean wa median ko prayog gari postal codes fill garna sakinna.)
##We also tried looking it up online but only using City and State is not enough to get a postal code since one city has multiple postal codes =>
#  (Online herera pani kosis gariyeko thiyo, tara City ra State matra ko adharma postal code pauna sakinna, kinabhane ekai city ma dherai postal codes hunchan.)
##So we remove the 11 rows with missing postal codes = (Hamlea 11 wota row remove gare jasko postal code cahi missing thiyo )

df_copy = df_copy.dropna()

#New shape after rows with missing values have been dropped =(Missing value bhayeka rows hataisakepachi data ko new shape banayeam )
df_copy.shape
```

```python
##Percent of rows removed after drooping rows with missing values = (Missing value bhayeka rows hataisakepachi hataiyeko rows ko percent dekhaye )

percent_rows_removed = ((df.shape[0] - df_copy.shape[0]) / df.shape[0]) * 100
print("Percent of rows removed using .dropna():", percent_rows_removed)
```

```python
##Checking for duplicate values (based on all columns besides Row ID) = (Row ID bhand bahira sabai columns dhrit duplicate values janch gardai)
duplicate_count = df_copy.duplicated(subset=['Order ID', 'Order Date', 'Ship Date', 'Ship Mode',
       'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State',
       'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category',
       'Product Name', 'Sales']).sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
##Checking which row is duplicated => (Kunai chai row dublicated xa ki check gardai )
duplicate_rows = df_copy[df_copy.duplicated(subset=['Order ID', 'Order Date', 'Ship Date', 'Ship Mode',
       'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State',
       'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category',
       'Product Name', 'Sales'], keep=False)]
duplicate_rows

##This could be a duplicate order or the customer could have bought multiple chairs so we did not remove this row
#  Yaha cahi dublicated order haru xa ra customer le cahi aautai saman dherai patak kineko huda hamlea cahi tyo row hatayenam
```

```python
#Checking datatypes for the columns = (Columns ko datatype haru janch gardai)
df.info()
```

Conversions required:  
1. Columns Order ID, Ship Mode, Customer ID, Customer Name, Segment, Country, City, State, Region, Product ID, Category, Sub-Category, Product Name are stored as 'object', so we convert to 'string'.
2. Order Date and Ship Date are stored as 'object', so we convert to 'datetime'.
3. Postal Code is stored as 'float', which we will change to 'string' since it is geographic data and keeping it numeric does not give us any meaning.

```python
##Changing columns Order ID, Ship Mode, Customer ID, Customer Name, Segment, Country, City, State, Region, Product ID, Category, Sub-Category, Product Name, Postal Code into string

str_columns = ['Order ID', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name', 'Postal Code']

for column in str_columns:
  df_copy[column] = df_copy[column].astype('string')
```

```python
#Changing order date and ship date from object to datetime. =(Order Date ra Ship Date lai object bata datetime ma changing  gardai)
df_copy['Order Date'] = pd.to_datetime(df_copy['Order Date'], format="%d/%m/%Y")
df_copy['Ship Date'] = pd.to_datetime(df_copy['Ship Date'], format="%d/%m/%Y")
```

## Step 3: Outlier Detection and Handling

```python
##Since our only numeric column is 'Sales', we will check for outliers in 'Sales' column using box plot and remove them using IQR method if necessary
# Hami sanga matra numeric column 'Sales' chha, tesaile 'Sales' column ma outliers cha ki chhaina bhanera box plot bata janch garne chau, ra awashyak parne ho bhane IQR method bata tyo outliers hataune chau.
plt.figure(figsize=(5, 12))
sns.boxplot(y=df_copy["Sales"])
plt.title("Box Plot for Sales")
plt.show()
```

```python
#We see extreme skewness so we remove outliers using the IQR method = (Hami le atyadhik skewness dekhe, tesaile IQR method ko madat le outliers hataune chau.)

Q1 = df_copy['Sales'].quantile(0.25)
Q3 = df_copy['Sales'].quantile(0.75)

iqr = Q3 - Q1
upper_limit = Q3 + 1.5 * iqr
lower_limit = Q1 - 1.5 * iqr

df_no_outliers = df_copy[(df_copy['Sales'] < upper_limit) & (df_copy['Sales'] > lower_limit)]
```

```python
#Checking the number of outliers removed =(Outliars ko number check gareko  aani teslai hatako  )
print("Number of outliers removed by IQR method:", df_copy.shape[0] - df_no_outliers.shape[0])
percent_removed = ((df_copy.shape[0] - df_no_outliers.shape[0])/df_copy.shape[0]) * 100
print("Percent of rows removed after outliers are removed:", percent_removed)
```

## Step 5: Feature Engineering

Steps in feature engineering:  
1. Create new feature.
2. Show visuals for this new feature.

```python
##Creating a new column for difference in ship date and order date
#Ship date ra order date ko antar dekhauune naya column create garne.
df_no_outliers['Ship Difference'] = df_no_outliers['Ship Date'] - df_no_outliers['Order Date']
df_no_outliers.head()
```

```python
##Grouping by the new 'Ship Difference' column to see the counts
# Naya 'Ship Difference' column anusar grouping garera counts herne.
sns.countplot(df_no_outliers['Ship Difference'], color = 'orange')
plt.title("Difference in ordering and shipping")
plt.xlabel("Number of Orders")
plt.ylabel("Ship Difference")
plt.show()

#We can also calculate the mean days it takes from ordering to shipping
# Ordering dekhi shipping samma lagne average din (mean days) pani calculate garna sakinchha.
mean_diff = df_no_outliers['Ship Difference'].mean()
#Converting it to number of dats
mean_days = mean_diff.total_seconds() / (60 * 60 * 24)
print("\nAverage number of days it takes from ordering to shipping:",mean_days)
```

```python
##Since our sales column has very extreme values (before removing outliers), we can create categorical bins based on different range
# Sales column ma dherai extreme values chan (outliers hataune aghi), tesai le haami different range ko adhar ma categorical bins create garna sakchhau.

#Checking minimum and maximum values of sales column
print("Maximum sales value:",df_copy['Sales'].max())
print("Minimum sales value:",df_copy['Sales'].min())
```

```python
#Based on the maximum and minimum sales values, we can create 5 different bins for the sales
df_copy['Sales Category'] = pd.cut(df_copy['Sales'], bins=[0, 100, 500, 1000, 5000, 23000], labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
df_copy.head()
```

```python
##Grouping by the new 'Sales Category' column to see the counts, sums and averages
sns.countplot(df_copy['Sales Category'], color = 'orange')
plt.title("Number of Orders by Sales Category")
plt.xlabel("Number of Orders")
plt.ylabel("Sales Category")
plt.show()

##Grouping to get the total sales for each sales category
total_sales = df_copy.groupby(['Sales Category'])[['Sales']].sum().reset_index().rename(columns={'Sales':'Total Sales'})

sns.barplot(y='Sales Category', x='Total Sales',data=total_sales, color = 'green')
plt.title("Total sales by Sales Category")
plt.xlabel("Total Sales")
plt.ylabel("Sales Category")
plt.show()

##Grouping to get the total sales for each sales category
avg_sales = df_copy.groupby(['Sales Category'])[['Sales']].mean().reset_index().rename(columns={'Sales':'Average Sales'})

sns.barplot(y='Sales Category', x='Average Sales',data=avg_sales, color = 'red')
plt.title("Average sales by Sales Category")
plt.xlabel("Average Sales")
plt.ylabel("Sales Category")
plt.show()

```
